# New Year Chaos
![New Year Chaos](New-Year-Chaos.png)

```python
for _ in range(int(input())):
    n=int(input())
    a=list(map(int,input().split()))
    if len(a)==1:
        print(a[0])
    else:
        d=0
        i=0
        while i<n:
            index_n=a.index(max(a))
            if index_n==len(a):
                a.remove(max(a))
            else:
                if (len(a)-index_n)<4:
                    print(a)
                    d+=(len(a)-index_n-1)
                    a.remove(max(a))
                else:
                    print('Too chaotic')
                    d=0
                    break
            i+=1
        if d!=0:
            print(d)
```

```python
for _ in range(int(input())):
    n=int(input())
    q=list(map(int,input().split()))   
    bribes = 0
    for i in range(len(q)-1,-1,-1):
        if q[i] - (i + 1) > 2:
            print('Too chaotic')
            bribes=0
            break
        for j in range(max(0, q[i] - 2),i):
            if q[j] > q[i]:
                bribes+=1
    if bribes!=0:
        print(bribes)
```

# Minimum-Swaps-2
![Minimum-Swaps-2](Minimum-Swaps-2.png)

```python
def minimumSwaps(arr):
    count = 0
    for i in range(len(arr)):
        while arr[i] != i+1:
            temp = arr[i];
            arr[i] = arr[temp-1];
            arr[temp-1] = temp;  
            count +=1;
    return count;
n=int(input())
a=list(map(int,input().split()))
print(minimumSwaps(a))
```
