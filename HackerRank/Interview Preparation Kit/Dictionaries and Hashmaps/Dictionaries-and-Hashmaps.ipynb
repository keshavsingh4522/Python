# Two Strings
<img src="Two-Strings.png" alt='Two-Strings.png' />

```python
for _ in range(int(input())):
    s1=input()
    s2=input()
    for i in s1:
        if i in s2:
            print('YES')
            break
    else:
        print('NO')
    print()
```

# Hash Tables: Ransom Note
![Hash Tables: Ransom Note](Hash-Tables--Ransom-Note.png)

```python
from collections import Counter as C
s1=input().split()
s2=input().split()
if C(s2)-C(s1)=={}:
    print('Yes')
else:
    print('No')
```
