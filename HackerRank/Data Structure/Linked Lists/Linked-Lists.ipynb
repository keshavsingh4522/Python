```python
def getNode(head, p):
    l=[]
    while head:
        l.append(head.data)
        head=head.next
    return l[-p-1]
```

```python
def getNode(head, p):
    tracked=head
    while head:
        if p>-1:
            p-=1
        else:
            tracked=tracked.next
        head=head.next
    return tracked.data
```

### [Cycle Detection](https://www.hackerrank.com/challenges/detect-whether-a-linked-list-contains-a-cycle/problem?h_r=next-challenge&h_v=zen&h_r=next-challenge&h_v=zen)


- **method 1 (hashing)**
- Time Complexity O(n)
- Auxiliary Space O(n)

```python
def has_cycle(head):
    l=set()
    while head:
        if head in l:
            return True
        else:
            l.add(head)
        head=head.next
    return False
```

- **method 2 (Floydâ€™s Cycle-Finding Algorithm)**
- Time Complexity O(n)
- Auxiliary Space O(1)

```python
def has_cycle(head):
    l1=head
    l2=head
    while l1 and l2 and l2.next:
        l1=l1.next
        l2=l2.next.next
        if l1==l2:
            return True
    return False
```
