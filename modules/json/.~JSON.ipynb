```python
# tuple---list
import json
x=(1,2,4)
print('before converting to json')
print(x)
print(type(x))
s=json.dumps(x)
print('after converting to json')
print(s)
print(type(s))
s1=json.loads(s)
print('from json to convert in actual data')
print(s1)
print(type(s1))
```

```python
# dict--dict
x={1:'a',2:'b',4:'c'}
print('before converting to json')
print(x)
print(type(x))
s=json.dumps(x)
print('after converting to json')
print(s)
print(type(s))
s1=json.loads(s)
print('from json to convert in actual data')
print(s1)
print(type(s1))
```

```python
x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}
s=json.dumps(x,indent=5,separators=(".","="),sort_keys=True)
# s1=json.loads(s)
print(s)
```

```python
data = {}
data['people'] = []
data['people'].append({
    'name': 'Scott',
    'website': 'stackabuse.com',
    'from': 'Nebraska'
})
data['people'].append({
    'name': 'Larry',
    'website': 'google.com',
    'from': 'Michigan'
})
data['people'].append({
    'name': 'Tim',
    'website': 'apple.com',
    'from': 'Alabama'
})
```

```python
print(json.dumps(data,indent=2,sort_keys=True,separators=("......","======")))
```

```python
# writing in json file
with open('data.json','w') as file:
    json.dump(data,file,indent=4)
```

```python
# reading from json file
import json
with open('data.json','r') as file:
    data=json.load(file)
    print(data)
    print(type(data))
```
